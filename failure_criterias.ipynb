```python
import numpy as np
import matplotlib.pyplot as plt
```

## Von misses criteria

```python
sigma_1 = np.linspace(-300, 300, 100)
sigma_2 = np.linspace(-300, 300, 100)

S1, S2 = np.meshgrid(sigma_1, sigma_2)    
```

```python
sigma_vm = np.sqrt(S1**2 - S1*S2 + S2**2)
sigma_adm = 200
```

```python
plt.figure(figsize=(6,6))

plt.contourf(S1, S2, sigma_vm, levels=50, cmap='jet')

plt.colorbar(label='Von Mises stress')

# Yield surface (VERY IMPORTANT)
plt.contour(S1, S2, sigma_vm, levels=[sigma_adm], colors='red', linewidths=2)

plt.xlabel('Sigma 1')
plt.ylabel('Sigma 2')
plt.title('Von Mises Yield Criterion')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.gca().set_aspect('equal')
plt.show()
```

```python
sigma_vm_clipped = np.clip(sigma_vm, None, sigma_adm)

plt.figure(figsize=(6,6))

plt.contourf(S1, S2, sigma_vm_clipped, levels=50, cmap='jet', vmin=0, vmax=sigma_adm)
plt.colorbar(label='Von Mises stress')

# Yield surface
plt.contour(S1, S2, sigma_vm, levels=[sigma_adm], colors='red', linewidths=2)

plt.xlabel('Sigma 1')
plt.ylabel('Sigma 2')
plt.title('Von Mises (clipped at admissible)')

plt.gca().set_aspect('equal')
plt.show()
```
