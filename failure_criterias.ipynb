```python
import numpy as np
import matplotlib.pyplot as plt
```

## Von misses criteria

```python
max_sigma = 300
sigma_adm = 220  

sigma_1 = np.linspace(-max_sigma, max_sigma, 100)
sigma_2 = np.linspace(-max_sigma, max_sigma, 100)

S1, S2 = np.meshgrid(sigma_1, sigma_2) 

sigma_vm = np.sqrt(S1**2 - S1*S2 + S2**2)
```

```python
plt.figure(figsize=(6,6))

cmap = plt.get_cmap('jet', 200)
cmap.set_over(cmap(np.linspace(0,1,256))[-1])  

# Definir niveles SOLO hasta 200
levels = np.linspace(0, sigma_adm, 50)

cf = plt.contourf(S1, S2, sigma_vm,
                  levels=levels,
                  cmap=cmap,
                  extend='max')  # 👈 activa la flecha

cbar = plt.colorbar(cf, label='Von Mises stress')

# Yield surface
plt.contour(S1, S2, sigma_vm,
            levels=[sigma_adm],
            colors='black',
            linewidths=4,
            alpha=.3)

plt.xlabel('Sigma 1')
plt.ylabel('Sigma 2')
plt.title('Von Mises Yield Criterion')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.gca().set_aspect('equal')
plt.show()
```
